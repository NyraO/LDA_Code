```python
import matplotlib.pyplot as plt
import pandas
import numpy as np

points = pandas.read_csv('points.csv')

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = points['alpha'].values
y = points['beta'].values
z = points['coherence'].values

#  x   y   z  
# 1.1,1.2,1.3
# 2.1,2.2,2.3
# 3.1,3.2,3.3
# 4.1,4.2,4.3

ax.scatter(x, y, z, c='r', marker='o')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas

points = pandas.read_csv('points.csv')

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.legend(['For K=4, Gibss model coherence for differet alpha and beta value'])

x = points['alpha'].values
y = points['beta'].values
z = points['coherence'].values

#  x   y   z  
# 1.1,1.2,1.3
# 2.1,2.2,2.3
# 3.1,3.2,3.3
# 4.1,4.2,4.3

# Function to map the colors as a list from the input list of x variables
def pltcolor(lst):
    cols=[]
    for l in lst:
        if l>=0.4:
            cols.append('red')
        elif l<0.2:
            cols.append('blue')
        else:
            cols.append('green')
    return cols
# Create the color


ax.scatter(x, y, z,c=pltcolor(z), marker='o')
ax.set_xlabel('alpha')
ax.set_ylabel('beta')
ax.set_zlabel('coherence')

plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas

df = pandas.read_csv('test7_header.csv')

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


df = df.loc[df['K'] == 5]
x = df['alpha'].values
y = df['beta'].values
z = df['coherence'].values
z = z.astype(np.float)
    

# Function to map the colors as a list from the input list of x variables
def pltcolor(lst):
    cols=[]
    for l in lst:
        if l>=0.4:
            cols.append('red')
        elif l<0.2:
            cols.append('blue')
        else:
            cols.append('green')
    return cols
# Create the color


ax.scatter(x, y, z,c=pltcolor(z), marker='o')
ax.set_xlabel('alpha')
ax.set_ylabel('beta')
ax.set_zlabel('coherence')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas
import numpy as np

points = pandas.read_csv('test7_header.csv')


def pltcolor(lst):
    cols=[]
    for l in lst:
        if l>=0.4:
            cols.append('red')
        elif l<0.2:
            cols.append('blue')
        else:
            cols.append('green')
    return cols

# Create the color
for k_ in range (2,11,1):  
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    df = points[points['K'] == k_]
    x = df['alpha'].values
    y = df['beta'].values
    z = df['coherence'].values
    z = z.astype(np.float)
    
    ax.scatter(x, y, z,c=pltcolor(z), marker='o')
    ax.set_xlabel('alpha')
    ax.set_ylabel('beta')
    ax.set_zlabel('coherence')

    plt.show()
```
